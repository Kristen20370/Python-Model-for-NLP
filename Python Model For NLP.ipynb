```python
import pandas as pd
from scipy.optimize import minimize
```

```python
df = pd.read_csv("Historical returns.csv")
df = df.transpose()
df
```

```python
cov_matrix=df.cov().copy()
cov_matrix
```

```python
ER = [0.04,0.07,0.18,0.06,0.07,0.17,0.29,0.48]
```

```python
n=len(df.columns)
#Assume the minimum expected return that the portfolio much achieve is e_0
e_0 = 0.015
#define objective function
def risk(w):
    risk_sum=0
    for i in range(n):
        for j in range(n):
            risk_sum +=w[i]*w[j]*cov_matrix.iloc[i,j]
    return risk_sum
#define minimum expected return inequality constraint
def constraint1(w):
    return sum([w[i]*ER[i] for i in range(n)])-e_0
#define weights sum up to 1 equality constraint
def constraint2(w):
    return sum([w[i] for i in range(n)])-1.0
#define the customer preference constraints
def constraint3(w):
    return w[0]+w[5]-0.3
def constraint4(w):
    return -w[6]-w[7]+0.15
def constraint5(w):
    return w[3]+w[5]-w[2]-w[6]

con1={'type':'ineq','fun':constraint1}
con2={'type':'eq','fun':constraint2}
con3={'type':'ineq','fun':constraint3}
con4={'type':'ineq','fun':constraint4}
con5={'type':'ineq','fun':constraint5}
cons=([con1,con2,con3,con4,con5])

#define bounds
b=(0,1)
bnds=(b,)*n

#define inital values for weights
w0=[0.2]*n

solution=minimize(risk,w0,method='SLSQP',bounds=bnds,constraints=cons)
```
